**Criando Gráfico scatter plot**

```python

df = pd.DataFrame({
    "paises": ["Brasil", "EUA", "China", "Alemanha", "Japão", "França", "Reino Unido", "Índia", "Rússia", "Canadá", "Austrália", "Coreia do Sul"],
    "investimento_1960": [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200],
    "investimento_2022": [500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000]
})


df_best_investments = df.loc[df["investimento_2022"] >= 1500]


plt.figure(figsize=(10, 6))
plt.scatter(df_best_investments["investimento_1960"], df_best_investments["investimento_2022"])
plt.title("Scatter Plot: Investimento (1960 vs 2022)")
plt.xlabel("Investimento em 1960")
plt.ylabel("Investimento em 2022")


for i, txt in enumerate(df_best_investments["paises"]):
    plt.annotate(txt, (df_best_investments["investimento_1960"].iloc[i], df_best_investments["investimento_2022"].iloc[i]))

plt.grid(True)
plt.show()
```
